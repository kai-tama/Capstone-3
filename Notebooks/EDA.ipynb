## Setting up for sentiment analysis

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.base import TransformerMixin
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from spacytextblob.spacytextblob import SpacyTextBlob
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')
```

```python
def get_sentiment(text):
    doc = nlp(text)
    polarity = doc._.blob.polarity
    return polarity
```

```python
reviews_df['polarity'] = reviews_df['body'].apply(get_sentiment)
```

```python
max(reviews_df.polarity)
```

```python
min(reviews_df.polarity)
```

I'm curious to see how much of an affect removing the stopwords will have on the polarity score and was pleasantly surprised!

```python
def remove_stopwords(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop]
    return ' '.join(tokens)
```

```python
reviews_df['cleaned_text'] = reviews_df['body'].apply(remove_stopwords)
```

```python
reviews_df['new_polarity'] = reviews_df['cleaned_text'].apply(get_sentiment)
```

```python
max(reviews_df.new_polarity)
```

```python
min(reviews_df.new_polarity)
```
