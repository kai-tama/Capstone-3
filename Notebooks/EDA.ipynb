## Word clouds from reviews

```python
from sklearn.pipeline import Pipeline
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from spacytextblob.spacytextblob import SpacyTextBlob
import matplotlib.pyplot as plt
import matplotlib as mpl
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

While there probably isn't too big of a difference, I still want to see the different results I get depending which 'stop words' I use.

```python
drop_words = ['musical', 'stage', 'broadway', 'play', 'show', 'theater', 'theatre', 'new york', 'york', 'photo by', 'guide', 
              'year', 'years', 'new', 'second', 'minute', 'hour', 'time', 'american', 'america', 'times', 'minutes', 'half',
             'hours', 'london']
stopwords = STOPWORDS.update(drop_words)
stop_words = STOP_WORDS.update(drop_words)
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')
```

```python
def remove_stopwords(text):
    doc = nlp(text)
    tokens = [token.text for token in doc if not token.is_stop]
    return ' '.join(tokens)
```

Before I generate the word clouds, I also want to get rid of peoples' names since those will definitely pop up a lot and aren't very relevant to understanding the sentiment of an article.

```python
def remove_people(text):
    doc2 = nlp(text)
    persons = [ent.text for ent in doc2.ents if not ent.label_ == 'PERSON']
    return ' '.join(persons)
```

```python
reviews_df['cleaned_text'] = reviews_df['body'].apply(remove_people)
```

```python
text = reviews_df.cleaned_text.tolist()

text = ' '.join(text).lower()
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wordcloud = WordCloud(stopwords=stop_words, background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

I tweak the paramaters a bit to see if I can get a better result.

```python
wordcloud = WordCloud(stopwords=stopwords, min_word_length=4, background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wordcloud = WordCloud(stopwords=stop_words, min_word_length=4, background_color='white').generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
