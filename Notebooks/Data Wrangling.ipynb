## Cleaning up the show title column, part 1

My Scrapy spider wasn't perfect in grabbing show titles from the articles, so I start by replacing any value in the show title column that contains ( with the article title, since it seems that if a show title wasn't grabbed, it's in the article title instead.

```python
for index, row in broadway_df.iterrows():
    if '(' in row['show_title']:
        broadway_df.loc[index, 'show_title'] = row['article_title']
```

```python
broadway_df.show_title.unique()
```

```python
for index, row in offbroadway_df.iterrows():
    if '(' in row['show_title']:
        offbroadway_df.loc[index, 'show_title'] = row['article_title']
```

Now we move on to the Playbill dataframe. I already know that I need to get rid of the dollar signs and commas, so I quickly take care of that.

### Playbill

```python
playbill_df['week_gross'] = playbill_df['week_gross'].str.replace('[$,]','', regex=True)
playbill_df['avg_ticket'] = playbill_df['avg_ticket'].str.replace('[$,]','', regex=True)
playbill_df['top_ticket'] = playbill_df['top_ticket'].str.replace('[$,]','', regex=True)
playbill_df['seats_sold'] = playbill_df['seats_sold'].str.replace(',','', regex=True)
playbill_df['seats_in_theater'] = playbill_df['seats_in_theater'].str.replace(',','', regex=True)
playbill_df['capacity_filled'] = playbill_df['capacity_filled'].str.replace('%','', regex=True)
```

```python
playbill_df
```

```python
playbill_df.dtypes
```

```python
playbill_df['week_gross'] = playbill_df['week_gross'].astype(float)
playbill_df['avg_ticket'] = playbill_df['avg_ticket'].astype(float)
playbill_df['top_ticket'] = playbill_df['top_ticket'].astype(float)
playbill_df['seats_sold'] = playbill_df['seats_sold'].astype(int)
playbill_df['seats_in_theater'] = playbill_df['seats_in_theater'].astype(int)
playbill_df['capacity_filled'] = playbill_df['capacity_filled'].astype(float)
```

```python
playbill_df.dtypes
```

Now that all of the columns have their appropriate types, I'm going to quickly group the dataframe by year and show name to ensure that there isn't anything that I have missed cleaning up.

```python
yearavg = playbill_df.groupby('year').mean(numeric_only=True)
yearavg
```

Looks like the top ticket price in 2021 either doesn't exist or my spider didn't scrape it properly - I hopped over to the Playbill site to take a look myself, and indeed there isn't any data for top ticket prices in 2021. I'll go ahead and fill that in with 0's for now.

```python
playbill_df['top_ticket'].fillna(0, inplace=True)
```

```python
yearavg = playbill_df.groupby('year').mean(numeric_only=True)
yearavg
```

I need to fix the values in the top_ticket column since the information either wasn't available or is lower than the average ticket price, which doesn't make sense.

```python
for index, row in playbill_df.iterrows():
    if row['avg_ticket'] > row['top_ticket']:
        playbill_df.at[index, 'top_ticket'] = row['avg_ticket']
```
